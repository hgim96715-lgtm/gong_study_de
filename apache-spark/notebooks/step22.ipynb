## streaming

```python
from pyspark.sql import (
    Row,
    SparkSession)
import pyspark.sql.functions as F
import pyspark.sql.types as t
```

```python
APP_NAME = "UniversalStreamingApp"
HOST = "localhost"
PORT = 9999
OUTPUT_MODE = "append" 

spark = SparkSession.builder.appName(APP_NAME).getOrCreate()

# 읽기 시작 readStream
raw_df=(
    spark
    .readStream
    .format("socket")
    .option("host",HOST)
    .option("port",PORT)
    .load()
)

processed_df=raw_df

# 쓰기 시작 writeStream
query=(
    processed_df
    .writeStream
    .outputMode(OUTPUT_MODE)
    .format("console")
    .option("truncate","false")
    .start()
)
query.awaitTermination()
```
