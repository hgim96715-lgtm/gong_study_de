## countByValue()

RDD에 있는 값들의 종류별 개수를 세어준다

- 반환값은 파이썬 딕셔너리
- map + reduceByKey+ collect를 한번에 해주는 Action 함수

```python
from pyspark.sql import SparkSession
```

```python
spark=(
    SparkSession
    .builder
    .appName("count-value")
    .master("spark://spark-master:7077")
    .getOrCreate()
)
```

```python
sc=spark.sparkContext
```

```python
text_file=sc.textFile("file:///workspace/data/grade.txt")
text_file.take(2)
```

```python
grade=text_file.map(lambda line:line.split(" ")[1])
# grade.take(2)
```

```python
grade_count=grade.countByValue()
grade_count
```

```python
# for grade,count in grade_count.items():
#     print(grade,count)
print(grade_count.items())
```

```python
# sort() vs sorted() 
# sorted : 원본은 그대로 두고, 새로운 복사본을 준다
# sort : 원본 변경
```

```python
for grade,count in sorted(grade_count.items(),key=lambda item:item[1],reverse=True):
    print(grade,":",count)
        
```
