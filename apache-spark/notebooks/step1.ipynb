## PySpark_Session_Context


- **SparkContext (`sc`):** (구식) RDD를 다루기 위한 원조 연결 통로. 엔진의 저수준 기능 제어.

- **SparkSession (`spark`):** (신식) RDD뿐만 아니라 DataFrame, SQL까지 모두 다루는 **통합 관리자**. (내부에 SparkContext를 포함하고 있음)


스파크 함수는 Transformation / Action 이 있다.

1. Transformation

- 결과값으로 새로운 RDD 생성
- 대표 함수 : map,filter,parallelize,distinct
- 계산하지 않는다

2. Action

- 결과값으로 데이터를 반환하거나 파일 저장
- 실제로 계산 하는 함수
- 대표 함수 : collect,count,take,saveAsTextFile,takeSample

```python
from pyspark.sql import SparkSession
```

```python
spark=(
    SparkSession
    .builder
    .appName("RDD-Practice")
    .master("spark://spark-master:7077")
    .getOrCreate()
)
```

```python
# 세션 객체 확인
print(spark)
```

```python
# 스파크 버전 확인(설치된 버전이 맞는지 확인)
print(spark.version)
```

```python
sc=spark.sparkContext
```

```python
big_rdd=sc.parallelize(range(100000))
```

```python
even_rdd=big_rdd.filter(lambda x:x%2==0)
```

```python
count=even_rdd.count()
```

```python
print(f"짝수의 개수:{count}")
```

```python
print(f"샘플데이터:{even_rdd.takeSample(False,5)}")
```

```python
spark.stop()
```
