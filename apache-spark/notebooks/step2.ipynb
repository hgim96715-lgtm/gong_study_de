## flatMap, reduceByKey, textFile

1. textFile()

- 텍스트 파일을 읽어서 줄(line)단위의 문자열 RDD로 만든다.
- 로컬 파일 읽기 `file://+절대경로`
- 절대경로에는 docker-compose.yml의 volumes 설정에서 만든것!

2. flatMap

| **구분** | **map (1:1)**                       | **flatMap (1:N)**              |
| ------ | ----------------------------------- | ------------------------------ |
| **비유** | 사과 1개 껍질 깎기 -> 깎인 사과 1개             | 문장 1개 쪼개기 -> 단어 10개            |
| **입력** | `["Hello World"]`                   | `["Hello World"]`              |
| **결과** | `[["Hello", "World"]]` (리스트 안에 리스트) | `["Hello", "World"]` (납작하게 펴짐) |
| **특징** | 개수가 변하지 않음                          | **개수가 늘어남 (뻥튀기)**              |

3. reduceByKey

- 데이터가 (key,value) 쌍으로 되어 있을때만 쓸 수 있다.

- 각자 자리에서 미리 합치고 나서 모으기때문에 데이터 이동량이 적음으로 빠르다

- groupByKey : 일단 다 모아놓고 계산 > 데이터 이동량이 많으므로 느리다



```python
from pyspark.sql import SparkSession
```

```python
spark=(
    SparkSession
    .builder
    .appName("Key-value")
    .master("spark://spark-master:7077")
    .getOrCreate()
)
```

```python
sc=spark.sparkContext
text_file=sc.textFile("file:///workspace/data/word.txt")
print(text_file.first())
```

```python
counts=(
    text_file
    .flatMap(lambda line: line.split(" "))
    .map(lambda word:(word,1))
    .reduceByKey(lambda a,b:a+b)
)
counts.take(2)
```

```python
# 터미널에서 확인가능
# counts.foreach(lambda x: print(x, flush=True))
```

```python
for datum in counts.collect():
    print(datum)
```
