## Pandas_Read_Write

- Read: pd.read_{포맷명}("파일명")
- 저장할때 : df.to_{포맷명}("파일명")

1. csv 읽기 read_csv

- header 주의 ! spark랑 다름
- spark :`header=True` 무조건 첫 줄이 header
- pandas : `header=0` 0번째 줄이 header, 없는 파일은 None, False가 아님!

2. 엑셀 읽기 read_excel
- openyxl 라이브러리가 설치해야한다

3. Parquet : pyarrow 라이브러리 필요

4. json : 웹 데이터나 로그 파일 읽을때 필요

- lines=True : 로그파일처럼 한줄에 Json하나씩 들어있는 경우 필수
