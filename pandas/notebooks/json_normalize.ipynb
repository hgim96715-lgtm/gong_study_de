## pandas json_mormalize

판다스가 제공하는 함수로,

딕셔너리나 리스트가 중첩된 복잡한 JSON데이터를 평평하게 만들어주는 것

**API 호출(`HttpOperator`) 결과**를 DB에 적재하기 전 가공할 때 필수적

```python
import pandas as pd
from pandas import json_normalize
import requests
```

```python
# 단일 JSON
```

```python
data={
    "name":"Luke Skywalker",
    "height": "172",
    "mass": "77"
}
df=json_normalize(data)
df
```

```python
# 리스트 JSON
```

```python
data = [
    {"name": "Luke", "height": "172", "mass": "77"},
    {"name": "Leia", "height": "150", "mass": "49"}
]
df=json_normalize(data)
df
```

```python
# 중첩된 dict JSON
# 점으로 연결된 새로운 컬럼이름으로 자동 생성
```

```python
data = [
    {
        "id": 1,
        "name": "Luke Skywalker",
        "location": {              
            "planet": "Tatooine",
            "galaxy": "Outer Rim"
        }
    }
]
df=pd.DataFrame(data)
df
```

```python
# 점으로 연결된 새로운 컬럼이름으로 자동 생성
df=json_normalize(data)
df
```

```python
# 리스트가 섞여있을때
#record_path : 이경로에 있는 리스트를 메인 데이터로 쓸때
# meta : 리스트 밖에 있는 정보도 옆에 붙일때
```

```python
data = {
    "name": "Luke",
    "films": [
        {"title": "A New Hope", "year": 1977},
        {"title": "Empire Strikes Back", "year": 1980}
    ]
}
```

```python
df=json_normalize(
    data,
    record_path=['films'],
    meta=['name']
)
df
```

```python
response=requests.get("https://swapi.dev/api/people/1/").json()
response
```

```python
df=json_normalize({
    "name":response["name"],
    "height":response["height"],
    "mass":response["mass"]
})
df
```
