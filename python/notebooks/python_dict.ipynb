### items/ keys /values

- 단, 인덱싱은 불가능하므로 순서대로 꺼내려면 list()로 변환

2. keys() : 키만 가져오기

4. values() : 값만 가져오기 

```python
config={"env":"prod","retries":3}
```

```python
type(config.items())
```

```python
config.items()
```

```python
for i,v in config.items():
    print(i,v)
```

```python
config.items()[0]
```

```python
list(config.items())[0]
```

```python
config.keys()
```

```python
config.values()
```

### 여러 예제들로 연습해보기

```python
dag_config = {
"owner": "team_A",
"retries": 2,
"email_on_failure": False
}


# 필수 키 목록
required_keys = ["owner", "retries", "start_date"]
```

```python
# 1. 없는 키가 무엇인지 찾아서 출력
```

```python
for key in required_keys:
    if key not in dag_config:
        print(key)
```

```python
# 없는 키를 리스트로 모으기

empty_keys=[]
for key in required_keys:
    if key not in dag_config:
        empty_keys.append(key)
print(empty_keys)
```

```python
# 리스트 컴프리헨션으로 변경해보기
empty_keys=[k for k in required_keys if k not in dag_config]
print(empty_keys)
```

```python
# 딕셔너리 ->리스트 변환 
```

```python
scores = {
"alice": 90,
"bob": 75,
"charlie": 85
}
```

```python
scores_list=list(scores.items())
print(scores_list)
```

#### sorted vs sort

- sorted : 원본 변경 안됨, 복사본 생성,새 리스트 반환, 딕셔너리 자체에는 직접 사용 불가 items()필요
- sort : 원본 변경 , 반환값 없음, 리스트 전용 매서드

```python
# 점수 기준으로 내림차순 정렬
grade=sorted(scores_list,key=lambda i:i[1],reverse=True)
print(grade)
```

```python
# 1등만 
top1=grade[0]
print(top1)
```

```python
# 상위 2명
top2=grade[:2]
print(top2)
```

```python
print(scores_list)
```

```python
my_list=[2,20,49,200,19]
result=my_list.sort()
print(result)
```
