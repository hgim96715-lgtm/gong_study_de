## python json

- 직렬화 : 파이썬 객체 -> json 문자열(전송, 저장용)
- 역직렬화 : json 문자열 -> 파이썬 객체(코드 사용용)

나중에 Airflow에서 HttpOperator로 받아온 결과는 텍스트이므로 json.loads()로 풀어서 쓴다.

Airflow XCom : 태스크 끼리 데이터를 주고 받을때 JSON 형태로 직렬화 해서 DB저장

s가 붙으면 문자열을 다룬다. `jsoqn.dumps(data)` , `json.loads(text)`

s가 없으면 파일을 다룬다 . `jsoqn.dump(data,f)`, `json.load(f)`

```python
import json
```

```python
# dict ->JSON(직렬화)
```

```python
info_data={
    "name":"Padmé Amidala",
    "height":"185",
    "mass":"45",
    "hair_color":"brown",
    "skin_color":"light",
    "eye_color":"brown",
    "birth_year":"46BBY",
    "gender":"female",
    "homeworld":"https://swapi.info/api/planets/8"
}
```

```python
print(type(info_data))
```

```python
json_str=json.dumps(info_data)
print(type(json_str))
```

```python
# JSON 문자열 -> Dict (역직렬화)
```

```python
restored_data=json.loads(json_str)
print("딕셔너리 변경,키로 접근가능:",restored_data['name'])
```

```python
# Dict -> 파일로 저장
```

```python
with open("../data/config.json","w") as f:
    json.dump(info_data,f)
```

```python
# 파일 -> Dict로 읽기
```

```python
with open("../data/config.json","r") as f:
    config=json.load(f)
print(f"파일 읽기:\n{config}")
```
