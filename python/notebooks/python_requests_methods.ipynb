## Requests Methods

- GET 은 서버에서 데이터를 보여줘 하고 가져오는 방식, 실행해도 서버 데이터 변하지 않음
- POST는 서버한테 데이터를 처리해줘 하고 보내는 방식, 서버에 새로운 데이터가 생기거나 변함

- GET : params=
- POST : data 또는 json=

GET, POST 둘다 Response객체를 반환한다

```python
import requests
```

```python
# get - 데이터 위치 URL
```

```python
payload = {'key1': 'value1', 'key2': 'value2'}
response=requests.get(
    "https://httpbin.org/get",
    params=payload
)
print(f"GET URL 확인, 주소뒤에 추가됨: {response.url}") 
print(response.json())
```

```python
# post - 새로운 데이터 생성
```

```python
payload={
    "username":"my_id",
    "password":"secret_pw"
}
response=requests.post(
    "https://httpbin.org/post",
    data=payload
)
print(f"POST URL 확인, 주소뒤에아무것도 안보임: {response.url}")
print(response.status_code)
print(response.json())
print(response.json()["form"])
```
