### 연습문제 
- 대문자 -> 소문자
- 소문자 -> 대문자 

```python
s="aBcDeFg"
```

```python
# 반복문 
res=""
for i in s:
    if i.isupper():
        res+=i.lower()
    else:
        res+=i.upper()
print(res)
```

```python
# swaocase()
print(s.swapcase())
```

```python
# re.sub
re.sub(r"[a-zA-Z]", lambda x: x.group(0).lower() if x.group(0).isupper() else x.group(0).upper(), s)
```
