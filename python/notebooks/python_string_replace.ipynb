### 연습문제 
- 대문자 -> 소문자
- 소문자 -> 대문자 

```python
s="aBcDeFg"
```

```python
# 반복문 
res=""
for i in s:
    if i.isupper():
        res+=i.lower()
    else:
        res+=i.upper()
print(res)
```

```python
# swaocase()
print(s.swapcase())
```

```python
# re.sub
re.sub(r"[a-zA-Z]", lambda x: x.group(0).lower() if x.group(0).isupper() else x.group(0).upper(), s)
```

#### 정규식함수 이해하기

- group()?
- 정규식함수를 그냥 출력하면 내용이 나오는게 아니라 Match ojbject를 반환
- 내용을 보려면 group()를 사용해야한다 

```python
import re
```

```python
text = "My number is 010-1234-5678"
pattern = r"\d{3}-\d{4}-\d{4}"
```

```python
m=re.search(pattern,text)
```

```python
print(m) # m은 문자열이 아니라 Mathch object
```

```python
print(type(m))
```

```python
print(m.group()) # 문자열 
```

```python
print(m.span()) # 위치 
```

```python
# 그럼 group()안에 숫자는 무엇일까?
```

```python
s = "010-1234-5678"

m = re.search(r"(\d{3})-(\d{4})-(\d{4})", s)
```

```python
print(m.group(0)) # 전체 
```

```python
print(m.group(1)) # 010
```

```python
print(m.group(2)) # 1234 앞번호 
```

```python
print(m.group(3)) # 5678 뒷번호
```
