## Reuests / Response

content-Type : 데이터 타입이거니깐 거기에 맞게 해석해달라는 것 , 서버의 안내서

- application/json : json 타입
- text/html : 일반적인 웹페이지
- image/jpeg or image/png : 이미지 데이터 , `.content`로 받아야한다.

requests.get() 을 실행하면 반환되는게 Response 객체

```python
# 전체 연습
```

```python
urls = {
    "json": "https://jsonplaceholder.typicode.com/todos/1",
    "html": "https://example.com",
    "image": "https://httpbin.org/image/png",
    "fail":"https://httpbin.org/status/500"
}
```

```python
for name,url in urls.items():
    # print(name,url)
    res=requests.get(url)
    # print(res)
    # print(f"{name},상태코드:{res.status_code}")
    if res.ok:
        # print("통신 성공!")
        content_type=res.headers.get("Content-type","")
        # print(f"데이터 타입: {name}-{content_type}")
        if "application/json" in content_type:
            data=res.json()
            print(f"{name}은 json데이터: {data['title']}")
            
        elif "text/html" in content_type:
            print(f"{name}은 html문서: \n {res.text[:100]}")

        elif "image" in content_type:
            with open("../data/image.jpg","wb") as f:
                f.write(res.content)
            print(f"{name}은 이미지 파일로 image.jpg파일로 저장함")

        else:
            print("알수 없는 파일입니다.")
    else:
        print(f"{name} 문제가 생겼습니다.")
```
