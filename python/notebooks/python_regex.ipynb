## Python Regex

- 파이썬 내장 모듈 import re를 반드시 선언해야 한다

**찾는 함수**

| 함수                       | 설명           | 반환 값                  | 특징                            |
| :----------------------- | :----------- | :-------------------- | :---------------------------- |
| **`re.search(패턴, 문자열)`** | **어디서든** 찾음  | Match Object          | **가장 추천!** 문장 중간에 있어도 찾음.     |
| `re.match(패턴, 문자열)`      | **맨 앞부터** 찾음 | Match Object          | 문장 중간에 있으면 못 찾음(None).        |
| `re.findall(패턴, 문자열)`    | **싹 다** 찾음   | **List** `['a', 'b']` | 결과가 리스트로 바로 나옴. (group 필요 없음) |


```python
import re
```

```python
text = "내 번호는 010-1234-5678 입니다"
pattern = r"\d{3}-\d{4}-\d{4}"
```

```python
print(f"search: {re.search(pattern,text)}")
```

```python
print(f"match: {re.match(pattern,text)}")
```

```python
print(f"findall:{re.findall(pattern,text)}\n type:{type(re.findall(pattern,text))}")
```

#### 정규식함수 이해하기

- group()?
- 정규식함수를 그냥 출력하면 내용이 나오는게 아니라 Match ojbject를 반환
- 내용을 보려면 group()를 사용해야한다
- 단, 괄호로 그룹으로 나눈것 ! 괄호를 잘해야한다!

```python
text = "My number is 010-1234-5678"
pattern = r"\d{3}-\d{4}-\d{4}"
```

```python
m=re.search(pattern,text)
```

```python
print(m) # m은 문자열이 아니라 Mathch object
```

```python
print(type(m))
```

```python
print(m.group()) # 문자열 
```

```python
# 그럼 group()안에 숫자는 무엇일까?
```

```python
s = "010-1234-5678"

m = re.search(r"(\d{3})-(\d{4})-(\d{4})", s)
```

```python
print(m.group(0)) # 전체 
```

```python
print(m.group(1)) # 010
```

```python
print(m.group(2)) # 1234 앞번호 
```

```python
print(m.group(3)) # 5678 뒷번호
```

```python
# print(m.group(4)) # 에러
```

```python
s = "a1b2c3"
m = re.search(r"\d", s)
print(m.group(0)) # 1
```

### 정규식 자주쓰는 패턴

| **기호** | **의미**       | **예시**                    |
| ------ | ------------ | ------------------------- |
| `\d`   | 숫자 (Digit)   | `\d{3}` (숫자 3개)           |
| `\w`   | 문자+숫자 (Word) | `\w+` (단어 1개 이상)          |
| `.`    | 아무 글자나 하나    | `.`                       |
| `*`    | 0개 이상 반복     | `a*` (없거나 a, aa, aaa...)  |
| `+`    | **1개 이상 반복** | `\d+` (숫자가 1개 이상 연속됨)     |
| `?`    | 있거나 없거나      | `https?` (http 또는 https)  |
| `^`    | 문자열의 시작      | `^Hello` (Hello로 시작하는 문장) |
| `$`    | 문자열의 끝       | `End$` (End로 끝나는 문장)      |

```python
# \d 숫자
text = "order123 number45 test"
re.findall(r"\d",text)
```

```python
# 숫자 3개 연속
re.findall(r"\d{3}",text)
# re.findall(r"\d{4}",text) # 안맞으면 []빈배열
```

```python
# \w 문자+숫자
text = "user_01 id-99 test"
re.findall(r"\w+", text)
```

```python
# 아무글자 하나 
text = "abc"
re.findall(r".", text)
```

```python
text = "a\nb"
re.findall(r".", text)
```

```python
# * 0개이상 반복, cb에는 a가 없는대도 매칭됨
text = "caaab a cb"
re.findall(r"ca*", text)
```

```python
# + 한개 이상반복, a가 꼭 있어야함!
text = "caaab a cb"
re.findall(r"ca+", text)
```

```python
# ? 있거나 없거나 
text = "http https ftp"
re.findall(r"https?", text)
```

```python
# ^ : 문자열의 시작 
text = "Hello world"
re.search(r"^Hello", text)
# re.search(r"^world", text) # 시작이 다르면 아무것도 반환되지 않는다
```

```python
# $ 문자열의 끝 
text = "This is the End"
re.search(r"End$", text)
# re.search(r"the$", text) # 끝이 다르면 아무것도 반환되지 않는다
```
