## HTTP 상태 코드

- 2xx : 성공
- 4xx : 클라이언트 에러 -> 내코드 수정 필요
- 5xx : 서버에러

`response.status_code` : 숫자를 직접 확인

`response.raise_for_status()`: 200번대가 아니면 강제로 에러 발생, 코드 중단 시키고 싶을때 유용

```python
import requests
```

```python
response=requests.get("https://jsonplaceholder.typicode.com/todos/1")

code=response.status_code

if code == 200:
    print("성공")
    data=response.json()

elif code == 404:
    print(" 실패: URL을 다시 확인하세요. 페이지가 없습니다.")

elif code == 403:
    print(" 권한 없음: API 키가 맞는지 확인하세요.")

elif code >= 500:
    print("서버 오류: 내 잘못 아님. 나중에 다시 시도 하세요.")

else:
    print(f"기타 상태 코드 발생 {code}")
```

```python
# raise_for_status사용
```

```python
try:
    response.raise_for_status()
    print("성공")
except requests.exceptions.HTTPError as e:
    print(f"에러 발생중단 : {e}")
```
