## Python_Functions

- def : 함수 정의 시작 키워드
- return : 함수의 결과 반환 (없다면 None반환)
- *args : 입력값들을 튜플로 묶어서 받음
- **kwargs : 입력값들을 딕셔너리로 묶어서 받음

```python
# return
```

```python
def add(a,b):
    result=a+b
    return result

val=add(3,5)
print(val)
```

```python
# *args : 갯수 무제한(튜플로 받음)
```

```python
def sum_all(*args):
    print(args)
    return sum(args)

print(sum_all(1,2,3))
print(sum_all(1,2,3,4,5,6,7,8))
```

```python
# **kwargs : 딕셔너리로 받음 
# Airflow할때 많이 사용
```

```python
def print_info(**kwargs):
    print(kwargs)

    if 'name' in kwargs:
        print(f"Hello, {kwargs['name']}")

print_info(name="Luke",age=25,planet="Tatooine")
```
