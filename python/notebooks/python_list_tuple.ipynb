## 리스트 수정 append insert pop extend

1. append : 맨 뒤에 추가
2. insert : 원하는 위치에 넣기
3. pop : 맨뒤에 제거
4. extend : 다른 리스트랑 합체 

```python
data=[1,2,3]
```

```python
data.append(4)
print(data)
```

```python
data.insert(1,99)
print(data)
```

```python
del_var=data.pop(0)
```

```python
nums=[100,200,300]
nums.append(del_var)
print(nums)
```

| 방식       | 결과        | 원본 변경 | 반환값   |
| -------- | --------- | ----- | ----- |
| `extend` | 요소 하나씩 추가 | ✅     | None  |
| `append` | 리스트째로 추가  | ✅     | None  |
| `+`      | 새 리스트 생성  | ❌     | 새 리스트 |


```python
a = [1, 2]
b = [3, 4]      
```

```python
a.append(b)
print(a)
```

```python
a.extend(b)
print(a)
```

```python
print(a+b)
```

### list comprehension

문법 : [ 변환식 for 변수 in 리스트 ]

조건 걸기 필터링 : [ 변환식 for 변수 in 리스트 if 조건 ]

if-else: [ (참일때 값) if 조건 else (거짓일때 값) for 변수 in 리스트 ]


```python
nums=[1,2,3,4]
a=[k*10 for k in nums ]
print(a)
```

```python
nums=list(range(11))
a=[k for k in nums if k%2==0]
a
```

```python
str_num=["even" if k %2==0 else "odd" for k in range(5)]
print(str_num)
```

```python
files = [
"data_01.csv",
"data_02.json",
"data_03.txt",
"log_01.csv",
"data_backup.json",
"backup_data.csv"
]
```

```python
# data_로 시작하는 파일만 고르기
start_data=[f for f in files if f.startswith("data_")]
print(start_data)
```

```python
# csv 또는 json 파일만
csv_json=[f for f in files if f.endswith((".csv",".json"))]
csv_json
```

```python
#data_로 시작하고,확장자가 .csv 또는 .json 인 파일

new_data=[f for f in files if (f.startswith("data") and f.endswith((".csv",".json")))]
new_data
```

```python
files = [
    "data_202401.csv",
    "data_202312.csv",
    "data_202402.json",
    "log_202401.csv",
    "data_202401.txt",
    "data_202403.parquet"
]
```

```python
#"2024"가 파일명에 포함되어 있고
#"data_"로 시작하며
#확장자가 .csv 또는 .json
```

```python
data_2024=[]
for f in files:
   if '2024' in f:
       data_2024.append(f)
print(data_2024)
```

```python
data_new_2024=[f for f in data_2024 if(f.startswith("data_") and f.endswith((".csv",".json")))]
print(data_new_2024)
```
